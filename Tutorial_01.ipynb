# Combinatorics and Probability



*   In a survey, 70\% of respondents prefer tea over coffee. If 10 people are randomly selected, what is the probability that exactly 7 prefer tea?



```python
def Comb(n,k):
  """
    This function calculates the combination of n and k.
  """
  return np.math.factorial(n)/(np.math.factorial(k)*np.math.factorial(n-k))

##start

# Probability of success (preferring tea)
p = 0.7

# Number of trials (people selected)
n = 10

# Number of successes (people preferring tea)
k = 7

probability = Comb(n,k)*(p**k)*((1-p)**(10))

print(f"The probability that exactly 7 people prefer tea is: {probability}")
##end
```




> Is there any other way to code the problem?








```python
##start

# Probability of success (preferring tea)
p = 0.7

# Number of trials (people selected)
n = 10

# Number of successes (people preferring tea)
k = 7

##end


# Calculate the probability using the binomial distribution
probability = binom.pmf(k, n, p)

print(f"The probability that exactly 7 people prefer tea is: {probability}")
```

```python
# Put all possible outcomes as lists in a dictionary:
outcomes = {}

i = 0

while True:
  i += 1
  outcomes[i] = []
  for l in range(i):
    # one means defective
    outcomes[i].append(1)
  for l in range(i,5):
    # zero means healty product
    outcomes[i].append(0)

## start

  # When we reach to the outcome which all are defected, then we need to break the loop
  if all(outcomes[i]) == None:
    break

## end

print("Possible Outcomes", outcomes)
```



*   How does this plot change by increasing the number of dices? (for example average of 10 dices instead of 2)



```python
# You don't need to change anything in this code box.
# Simply run the code to see the plots.

def Experiment(numbers_of_dice,n_trials):

  N = n_trials
  sum = np.random.randint(low = 1, high= 7, size= N, dtype=int)
  for i in tqdm(range(numbers_of_dice - 1)):
    sum += np.random.randint(low = 1, high= 7, size= N, dtype=int)
  x = sum/numbers_of_dice

  return x

n_trials = 10000
numbers_list = [2, 3]
fig1, axes = plt.subplots(1, 2)
for i, n in enumerate(numbers_list):
    ax = axes[i]
    x = Experiment(n, n_trials)
    unique, counts = np.unique(x, return_counts=True)
    ax.scatter(unique, counts/n_trials)
    ax.set_xlabel("possible outcome")
    if i == 0:
        ax.set_ylabel("frequency")
    ax.set_title(f"number of different \n dices = {n}")
    plt.subplots_adjust(wspace=0.4)
    ax.set_xlim([0,7])

plt.yticks([])
plt.show()
plt.close(fig1)

n_trials = 10000
numbers_list = [10, 100]
fig2, axes = plt.subplots(1, 2)
for i, n in enumerate(numbers_list):
    ax = axes[i]
    x = Experiment(n, n_trials)
    unique, counts = np.unique(x, return_counts=True)
    ax.scatter(unique, counts/n_trials)
    ax.set_xlabel("possible outcome")
    if i == 0:
        ax.set_ylabel("frequency")
    ax.set_title(f"number of different \n dices = {n}")
    plt.subplots_adjust(wspace=0.4)
    ax.set_xlim([0,7])

plt.yticks([])
plt.show()
plt.close(fig2)

n_trials = 10000
numbers_list = [1000, 10000]
fig3, axes = plt.subplots(1, 2)
for i, n in enumerate(numbers_list):
    ax = axes[i]
    x = Experiment(n, n_trials)
    unique, counts = np.unique(x, return_counts=True)
    ax.scatter(unique, counts/n_trials)
    ax.set_xlabel("possible outcome")
    if i == 0:
        ax.set_ylabel("frequency")
    ax.set_title(f"number of different \n dices = {n}")
    plt.subplots_adjust(wspace=0.4)
    ax.set_xlim([0,7])

plt.yticks([])
plt.show()
plt.close(fig3)
```
